```python
#cafe management system
name=input('enter your name:')
print(f" Hello!{name},How may I Assist You ")
print("What would you like to order")
menu = {
    "coffee": 10,
    "tea": 25,
    "maggie": 30,
    "sandwich": 45,
    "soda": 20
}

print("\n--- Our Menu ---")
for item, price in menu.items():
    print(f"{item.capitalize()}: ${price:.2f}")
print("------------------")

order_list = []
total_bill = 0

while True:
    item_name = input("\nEnter the item you want to order (or 'done' to finish): ").lower()
    if item_name == 'done':
        break
    elif item_name in menu:
        order_list.append(item_name)
        total_bill += menu[item_name]
        print(f"{item_name.capitalize()} added to your order.")
    else:
        print("Sorry, that item is not on our menu. Please choose from the available items.")

if order_list:
    print("\n--- Your Order ---")
    for item in order_list:
        print(f"- {item.capitalize()}")
    print(f"------------------")
    print(f"Your total bill is: ${total_bill:.2f}")
    print("Thank you for your order!")
else:
    print("You didn't order anything. Goodbye!")
```
