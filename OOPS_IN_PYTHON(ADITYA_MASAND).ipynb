```python
class BankAccount:
  def __init__(self, name, account_number, initial_balance):
    self.name = name
    self.account_number = account_number
    self.balance = initial_balance

  def deposit(self, amount):
    print(f"Depositing: {amount}")
    self.balance += amount
    print("Balance updated successfully")
    print("Updated balance is:", self.balance)

  def withdraw(self, amount):
    print(f"Attempting to withdraw: {amount}")
    if amount > self.balance:
       print("Insufficient balance")
    else:
      self.balance -= amount
      print("Balance updated successfully")
      print("Updated balance is:", self.balance)
b1 = BankAccount("Aditya", "ACC123", 5000)
b1.deposit(1000)
b1.withdraw(10000)
b1.withdraw(500)
```

```python
class Vehicle:
   def __init__(self, brand, model):
       self.brand = brand
       self.model = model

   def display(self):
       print(f"Vehicle Brand: {self.brand}, Model: {self.model}")

class Car(Vehicle):
   def __init__(self, brand, model, fueltype):
       super().__init__(brand, model)
       self.fueltype = fueltype

   def display(self):
       super().display()
       print("Fuel Type is:", self.fueltype)

class Bike(Vehicle):
    def __init__(self, brand, model, enginecapacity):
        super().__init__(brand, model)
        self.enginecapacity = enginecapacity

    def display(self):
        super().display()
        print("Engine Capacity is:", self.enginecapacity)
v = Vehicle('KIA', 'SUV')
v.display()
c = Car('Tesla', 'Model 3', 'Electric')
c.display()
b = Bike('Yamaha', 'R1', '1000cc')
b.display()
```
