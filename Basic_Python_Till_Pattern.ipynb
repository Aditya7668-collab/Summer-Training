```python
#pythagorean triplets
a=int(input("enter first number:"))
b=int(input("enter second number:"))
c=int(input("enter third number:"))
if(a*a+b*b==c*c):
    print("pythogorean triplets are:",a,b,c)
else:
    print("pythogorean triplets do not exist")
```

```python
# prime numbers between 1 to 100
print("Prime numbers between 1 and 100:")
for i in range(2, 101):
    is_prime = True
    for j in range(2, int(i**0.5) + 1):
        if i % j == 0:
            is_prime = False
            break
    if is_prime:
        print(i, end=" ")
```

```python
#difference between si and ci if p=1000 and r=10% and t=3
p=1000
r=10
t=3
si=(p*r*t)/100
ci=p*(pow(1+(r/100),t))-p
print(si)
print(ci)
print("difference between si and ci is:",ci-si)
```

```python
#find area and perimeter of a circle
pi=3.14
r=int(input('enter radius of a circle:'))
area=pi*r*r
perimeter=2*pi*r
print("area is :",area)
print("perimeter is:",perimeter)
```

```python
#find diagonal of a rectangle with l=3 and b=5
l=3
b=5
diagonal=(l*l+b*b)**0.5
print("diagonal is:",diagonal)
```

```python
#print pattern
#*
#**
#***
#****
#*****
#******
#*****
#****
#***
#**
#*
for i in range (1,12):
  if(i<=6):
    for j in range(1,i+1):
        print("*",end="")
  else:
       for j in range(1,13-i):
        print("*",end="")
  print("\n")
```
