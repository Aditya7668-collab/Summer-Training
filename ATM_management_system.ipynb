````python

name=input("enter your name:")
print(f"Hello {name} welcome to the atm")
print("enter your atm card")
def selectlanguage(language):
    if language=="Hindi":
       print("you have selected Hindi as your language")
    elif language=="English":
       print ("you have selected English as your language")
    else:
      print("you have selected a wrong language! pls try again")
def checkbalance(atm_pin_correct, balance, atm_pin_inputted):
    if atm_pin_inputted!=atm_pin_correct:
       print("you have entered the wrong atm pin")
    elif atm_pin_inputted==atm_pin_correct:
      print(f"your balance is {balance}")
def cashwithdrawal(amount, balance):
    if amount > balance:
      print("not enough balance in your account")
    else:
      print(f"Cash of ${amount} withdrawed successfully. Remaining balance: ${balance - amount}")
      return balance - amount
    return balance
def pinchange(current_atm_pin):
    print("enter your new atm pin")
    new_pin=int(input("enter your new pin"))
    print("pin changed successfully")
    return new_pin # Return the new pin

#----- ATM MANAGEMENT SYSTEM ---------
current_atm_pin=1234
current_balance=10000

while True: # Loop indefinitely until user chooses to exit
  print("\n--- ATM Menu ---")
  print("1. Select Language")
  print("2. Check Balance")
  print("3. Cash Withdrawal")
  print("4. Pin Change")
  print("0. Exit")

  try:
    choice=int(input("Enter your choice: "))
  except ValueError:
    print("Invalid input. Please enter a number.")
    continue

  if choice==1:
    language=input("Enter your language (Hindi/English): ")
    selectlanguage(language)
  elif choice==2:
    atm_pin_inputted=int(input("Enter your atm pin: "))
    checkbalance(current_atm_pin, current_balance, atm_pin_inputted)
  elif choice==3:
    atm_pin_inputted=int(input("Enter your atm pin to withdraw: "))
    if atm_pin_inputted == current_atm_pin:
        amount=int(input("Enter the amount you want to withdraw: "))
        current_balance = cashwithdrawal(amount, current_balance)
    else:
        print("Incorrect PIN. Withdrawal failed.")
  elif choice==4:
    atm_pin_inputted=int(input("Enter your current atm pin to change: "))
    if atm_pin_inputted == current_atm_pin:
        current_atm_pin = pinchange(current_atm_pin)
    else:
        print("Incorrect PIN. Pin change failed.")
  elif choice==0:
    print("Thank you for using our ATM!")
    break # Exit the loop
  else:
    print("Invalid choice. Please select a valid option.")

print("Take out your atm card.")
````
